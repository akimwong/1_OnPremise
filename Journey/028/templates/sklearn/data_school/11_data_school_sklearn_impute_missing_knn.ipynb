```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder
from sklearn.compose import make_column_transformer, ColumnTransformer, make_column_selector
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.experimental import enable_iterative_imputer   # <--------- 1.
from sklearn.impute import IterativeImputer   # <--------- 2.
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split 
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
train = pd.read_csv(r'./dataset/train.csv', sep=';')
```

```python
X = train.drop(['PassengerId', 'Name', 'Ticket', 'Sex', 
                'Parch', 'Ticket', 'Cabin','Embarked','Survived'], axis=1)
X = X.iloc[12:35]
X
```

```python
# experimental imputer, para comparar imputaciones con knn y SimpleImputer
impute_it = IterativeImputer()
impute_it.fit_transform(X)
```

```python
# knn sólo permite imputar con variables NUMÉRICAS
impute_knn = KNNImputer (n_neighbors = 4)
impute_knn.fit_transform(X)
```

```python
# utiliza la media para hacer todass las imputaciones
impute_simple = SimpleImputer()
impute_simple.fit_transform(X)
```
