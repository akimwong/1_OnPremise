```python
import random
def calcularGanancias(retweets):
    ganancia = retweets* random.randint(3,5)
    return ganancia
df["ganancias"] = df["retweets"].apply(calcularGanancias)

df.head(5)
```

```python
df = pd.read_csv(url, index_col=0)
```

```python
def calculaPopularidad(fila):
    popular = fila["followees"]/fila["followers"]
    return popular
df["popularidad"] = df.apply(calculaPopularidad, axis=1)
df.head(5)
```
