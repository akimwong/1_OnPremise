## 1. Ventana con un botón que imprime un mensaje fuera de la ventana

```python
import tkinter
import customtkinter

customtkinter.set_appearance_mode("dark")  # Modes: system (default), light, dark
customtkinter.set_default_color_theme("blue")  # Themes: blue (default), dark-blue, green

app = customtkinter.CTk()  # crea una ventana customtkinter
app.geometry("400x240")    # define el tamaño 'por defecto' de la ventana
app.resizable(width=False, height=False)  # indica que no pueden cambiarse las dimensiones de la ventana

def button_function():     # función que imprime un mensaje
    print("button pressed")  

# Creación de botón que está asociado a función button_function()
button = customtkinter.CTkButton(master=app, text="Preparar email", command=button_function)
# Define ubicación del botón dentro de la ventana
button.place(relx=0.5, rely=0.5, anchor=tkinter.CENTER)

app.mainloop()
```

## 2. Ventana con un botón que imprime un mensaje fuera de la ventana ctk

```python
import customtkinter


class App(customtkinter.CTk):
    def __init__(self):
        super().__init__()
        self.geometry("600x500")
        self.title("CTk example")

        # add widgets to app
        self.button = customtkinter.CTkButton(self, command=self.button_click)
        self.button.grid(row=0, column=0, padx=20, pady=10)

    # add methods to app
    def button_click(self):
        print("button click")


app = App()
app.mainloop()
```

## 3. Ventana con primer combobox

```python
import customtkinter


class App(customtkinter.CTk):
    def __init__(self):
        super().__init__()

        self.geometry("500x300")
        self.title("small example app")
        self.minsize(300, 200)

        # create 2x2 grid system
        self.grid_rowconfigure(0, weight=1)
        self.grid_columnconfigure((0, 1), weight=1)

        self.textbox = customtkinter.CTkTextbox(master=self)
        self.textbox.grid(row=0, column=0, columnspan=2, padx=20, pady=(20, 0), sticky="nsew")

        self.combobox = customtkinter.CTkComboBox(master=self, values=["S.TA", "Z.TO","B.SA","B.ND"])
        self.combobox.grid(row=1, column=0, padx=20, pady=20, sticky="ew")
        self.button = customtkinter.CTkButton(master=self, command=self.button_callback, text="Insert Text")
        self.button.grid(row=1, column=1, padx=20, pady=20, sticky="ew")

    def button_callback(self):
        self.textbox.insert("insert", self.combobox.get() + "\n")


if __name__ == "__main__":
    app = App()
    app.mainloop()
```

## 4. Ventana con 3 desplegables nombrando cada uno de ellos

```python
import pandas as pd
import customtkinter    # Importa la biblioteca personalizada customtkinter
from openpyxl import load_workbook

customtkinter.set_appearance_mode("dark")  # Modes: system (default), light, dark
customtkinter.set_default_color_theme("blue")  # Themes: blue (default), dark-blue, green

# Define un diccionario que asocia cada opción del menú desplegable con el nombre del archivo Excel correspondiente
#file_dict = {"S.TA": "archivo_STA.xlsx", "Z.TO": "archivo_ZTO.xlsx", "B.SA": "archivo_BSA.xlsx", "B.ND": "archivo_BND.xlsx"}


class App(customtkinter.CTk):    # Define una clase llamada "App" que hereda de la clase "CTk" de customtkinter
    def __init__(self):    # Define el método constructor de la clase "App"
        super().__init__()    # Inicializa la superclase "CTk"

        # Configura la geometría de la ventana, el título, y el tamaño mínimo
        self.geometry("600x300")
        self.title("faroCGM")
        self.minsize(300, 200)

        # Crea un sistema de rejilla de 2x2
        self.grid_rowconfigure(1, weight=1)
        #self.grid_columnconfigure(3, weight=1)
        self.grid_columnconfigure((0, 3), weight=1)

        # Crea una caja de texto y la coloca en la ventana
        self.textbox = customtkinter.CTkTextbox(master=self)
        self.textbox.grid(row=0, column=0, columnspan=2, padx=20, pady=(20, 0), sticky="nsew")

        # Crea un menú desplegable con opciones y lo coloca en la ventana
        self.combobox1 = customtkinter.CTkComboBox(master=self, values=["S.TA", "Z.TO"])
        self.combobox1.grid(row=1, column=0, padx=20, pady=20, sticky="ew")
        self.combobox1label = customtkinter.CTkLabel(master=self, text="Central")
        self.combobox1label.grid(row=1, column=0, columnspan=1, padx=20, pady=0, sticky="n")

        self.combobox2 = customtkinter.CTkComboBox(master=self, values=["Todas", "Pendientes"])
        self.combobox2.grid(row=1, column=1, padx=20, pady=20, sticky="ew")
        self.combobox2label = customtkinter.CTkLabel(master=self, text="En caliente", fg_color="transparent")
        self.combobox2label.grid(row=1, column=1, columnspan=1, padx=20, pady=0, sticky="n")

        self.combobox3 = customtkinter.CTkComboBox(master=self, values=[])
        self.combobox3.grid(row=1, column=2, padx=20, pady=20, sticky="ew")
        self.combobox3label = customtkinter.CTkLabel(master=self, text="Sgipe")
        self.combobox3label.grid(row=1, column=2, columnspan=1, padx=20, pady=0, sticky="n")
        
        # Crea un botón y lo coloca en la ventana, y define la función que se ejecutará cuando se haga clic en el botón
        self.button = customtkinter.CTkButton(master=self, command=self.button_callback, text="Email")
        self.button.grid(row=1, column=3, padx=20, pady=20, sticky="ew")

    def button_callback(self):
        filename_dict = {"S.TA": "CdM_Seguimiento_CCVV_HUAWEI_S.TA.xlsx", "Z.TO": "CdM_Seguimiento_CCVV_ERICSSON_Z.TO.xlsx"}  # Define un diccionario que asocia cada opción del menú desplegable con el nombre del archivo Excel correspondiente
        filename = filename_dict.get(self.combobox1.get())  # Obtiene el nombre del archivo Excel correspondiente a la opción seleccionada en el combobox1
        if filename:
            try:
                df = pd.read_excel(filename, header=1, sheet_name='RI_FO')  # Lee el archivo de Excel y carga los datos en un dataframe de pandas
                
                # Aplica el primer filtro: registros con TipoDeCorte "Corte en caliente"
                df = df[df["TipoDeCorte"] == "Corte en caliente"]
                
                # Aplica el segundo filtro: registros con FechaSolicitudOA vacía o no vacía, según la opción seleccionada en el combobox2
                if self.combobox2.get() == "Todas":
                    df = df[df["SGIPE_Actuación"].notna()]
                elif self.combobox2.get() == "Pendientes":
                    df = df[(df["SGIPE_Actuación"].notna()) & (df["FechaSolicitudOA"].isna())]                
                
                # Obtiene los valores únicos de la columna "SGIPE_Actuación" y los ordena de forma ascendente
                #sgipe_actuacion = df["SGIPE_Actuación"].drop_duplicates().sort_values(ascending=True)
                sgipe_actuacion_values = df["SGIPE_Actuación"].dropna().unique().tolist()
            
                # Actualiza los valores del combobox3 con los valores de "sgipe_actuacion"
                #self.combobox3.set_values(sgipe_actuacion.to_list())
                self.combobox3.set_values(sgipe_actuacion_values)
                #self.combobox3 = customtkinter.CTkComboBox(master=self)
                #self.combobox3.grid(row=1, column=2, padx=20, pady=20, sticky="ew")                
                
                self.textbox.delete("1.0", "end")  # Borra el contenido actual de la caja de texto
                self.textbox.insert("end", str(df))  # Muestra los datos del dataframe
                
            except FileNotFoundError:
                self.textbox.delete("1.0", "end")  # Borra el contenido actual de la caja de texto
                self.textbox.insert("end", "No se encontró el archivo " + filename)  # Muestra un mensaje de error si el archivo no se encuentra
        else:
            self.textbox.delete("1.0", "end")  # Borra el contenido actual de la caja de texto
            self.textbox.insert("end", "Seleccione una opción válida del menú desplegable")  # Muestra un mensaje de error si la opción seleccionada no está en el diccionario

        
    def button_callback(self):    # Define la función que se ejecutará cuando se haga clic en el botón
        self.textbox.insert("insert", self.combobox1.get() + "\n")    # Agrega el texto seleccionado del menú desplegable a la caja de texto

if __name__ == "__main__":
    app = App()    # Crea una instancia de la clase "App"
    app.mainloop()    # Inicia el ciclo principal del programa, que mantiene la ventana abierta y a la espera de interacciones del usuario
```
